```python
# 필요한 모듈을 import 하고 데이터를 df 라는 변수에 지정하세요
# Nan 값은 0으로 채워져야 하며 아래와 같은 내용으로 출력되어야 합니다. 
# 1970년 ~ 2021년 까지의 데이터 또는 1970 ~ 2022 년(누적데이터) 로 컬럼을 변경 하세요 

df= pd.read_csv("시도별 전출입 인구수.xlsx - 데이터.csv", thousands=',')
```

```python
# 맨위 5줄만 표시 하시오 
df = df.fillna(0)
df
```

```python
# 전출지가 서울인 사람이 서울이 아닌곳으로 전입한 열을 선택하여 df2에 저장하세요 
# 서울(전출) -> 타지역(전입)
df2= df[df["전출지별"]=="서울특별시"]
```

```python
df2
```

```python
# 컬럼을 전출지별, 전입지별, 항목, 1970 년, 1971년, ..., 2022. 10월 로 정렬 하시오
```

```python
# 전출지별과 항목 컬럼을 삭제하고, index를 전입지 별로 지정하시오 
del df2["전출지별"]
```

```python
df2 = df2.set_index(["전입지별"])
df2
```

```python
df2.info()
```

```python
# 서울에서 경기도로 이동한 인구 데이터를 df_su2ggd에 저장하세요. 
df_su2ggd = df2.loc["경기도"]
```

```python
df_su2ggd
```

```python
# 서울에서 경기도로 이동한 인구를 plot으로 나타내 보세요 
df_su2ggd = df_su2ggd.str.replace(',','').astype('int64')
df_su2ggd.head()
```

- Matplotlib 스타일 리스트 출력

```python
# 서울에서 경기도로 이동한 데이터와 서울에서 부산으로 이동한 인구를 동시에 plot에 표현해 보세요 
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] =False

plt.title("서울 -> 경기")
df_su2ggd.plot()
plt.show
```

# 막대그래프

```python
# 2000년 이후로 서울에서 ['부산광역시','대구광역시','인천광역시']로 이동한 인구를 막대그래프를 이용해 그려보세요 
df2=df2.loc[:,'2000':]
df2.head()
```

```python
df_su2bs = df2.loc['부산광역시']
df_su2dg = df2.loc['대구광역시']
df_su2ic = df2.loc['인천광역시']

df_su2dg
```

```python
df_su2bs.info()
```

```python
df_su2bs = df_su2bs.str.replace(',','').astype('int64')
df_su2dg = df_su2dg.str.replace(',','').astype('int64')
df_su2ic = df_su2ic.str.replace(',','').astype('int64')
```

```python
df_su2bs.plot()
```

```python
df_su2bs.values
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df_su2ic.index,df_su2ic.values,label = "인천광역시")
plt.bar(df_su2bs.index,df_su2bs.values,label="부산광역시")
plt.bar(df_su2dg.index,df_su2dg.values,label="대구광역시")


plt.legend( title='지역')
```
